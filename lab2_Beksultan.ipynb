```python
df1 = df[df['Total Population'] != '-']
df1
```

```python
print(df1.describe())
print(df1.max())
print(df1.min())
print(df1["Total Population"].describe())
```

```python
#2.8.	Выполните сортировку данных по нескольким критериям
df2 = df1.sort_values(by=['Year', "Country"])
print(df2)
```

```python
df1Fl = df1
df1Fl["Total Population"] = df1Fl["Total Population"].str.replace(',', "").astype(float)
df1Fl
```

```python

mx = df1Fl.sort_values(by=["Total Population"])
print("Название страны, которая имеет максимальную численность населения =",mx["Country"].iloc[-1])
print("Название страны, которая имеет минимальную численность населения =",mx["Country"].iloc[0])
```
